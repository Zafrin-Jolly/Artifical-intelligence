# BFS

```python
romanian_map = {
    'Arad': {'Sibiu':140, 'Zerind':75, 'Timisoara':118},
    'Zerind': {'Arad':75, 'Oradea':71},
    'Oradea': {'Zerind':71, 'Sibiu': 151},
    'Sibiu': {'Arad':140, 'Oradea':151, 'Fagaras':99, 'Rimnicu':80},
    'Timisoara': {'Arad': 118, 'Lugoj':111},
    'Lugoj': {'Timisoara':111, 'Mehadia':70},
    'Mehadia': {'Lugoj': 70, 'Drobeta':75},
    'Drobeta': {'Mehadia':75, 'Craiova':120},
    'Craiova': {'Drobeta': 120, 'Rimnicu': 146, 'Pitesti': 138},
    'Rimnicu': {'Sibiu': 80, 'Craiova': 146, 'Pitesti': 97},
    'Fagaras': {'Sibiu':99, 'Bucharest':211},
    'Pitesti': {'Rimnicu': 97, 'Craiova':138, 'Bucharest':101},
    'Bucharest': {'Fagaras':211, 'Pitesti': 101, 'Giurgiu':90, 'Urziceni':85},
    'Giurgiu': {'Bucharest': 90},
    'Urziceni': {'Bucharest': 85, 'Vaslui':142, 'Hirsova':98},
    'Hirsova': {'Urziceni':98, 'Eforie':86},
    'Eforie': {'Hirsova':86},
    'Vaslui': {'Iasi': 92, 'Urziceni':142},
    'Iasi': {'Vaslui':92, 'Neamt': 87},
    'Neamt': {'Iasi':87}
}
```

```python
from queue import Queue


def bfs(startingNode, destinationNode):
    visited = {}#to store explored nodes
    distance = {} #distance from node to node
    parent = {}#to store path

    bfs_traversal_output = []#to store the order of traversal
    frontier = Queue() #the non-explored nodes

    for city in romanian_map.keys(): #by default
        visited[city] = False
        parent[city] = None
        distance[city] = -1

    startingCity = startingNode #considering the cities as nodes
    visited[startingCity] = True #first starting node/city visited
    distance[startingCity] = 0 #distance from starting node to starting node is obviously zero
    frontier.put(startingCity)    # arad 
    found = False

    while not frontier.empty(): 
        u = frontier.get()     # in the first iteration it will be arad
        bfs_traversal_output.append(u) 

        for v in romanian_map[u].keys():
            if not visited[v]:                
                visited[v] = True 
                parent[v] = u #parent node,where v was expanded from
                distance[v] = distance[u] + romanian_map[u][v]  #to find the distance of v from goal node,adding the distance of u and distance from u to v             
                print("expanding "+parent[v]+" and got "+v) 
                frontier.put(v) 
                if v== destinationNode: 
                    found =True #checking if v is the destination node
                    print("\nSo the destination node "+v+" is found!")
                    print("\n")
                    break
                
        if found:
            
            break
            
    g = destinationNode #defining goal node
    path = [] 
    
    while g is not None:
        path.append(g) 
        g = parent[g]
    
    path.reverse() #reverse to count the full distance from starting node to destination node
    print(path)
    print(distance[destinationNode])
    print(bfs_traversal_output)

bfs('Arad', 'Bucharest')
```

# Uniform Cost Search

```python
from queue import PriorityQueue

def ucs(startingNode, destinationNode):
    visited = {}
    distance = {}
    parent = {}

    ucs_traversal_output = []
    pq = PriorityQueue() #which nodes needs tobe explored first

    for city in romanian_map.keys():
        visited[city] = False
        parent[city] = None
        distance[city] = -1

    startingCity = startingNode
    visited[startingCity] = True
    distance[startingCity] = 0
    pq.put((0,startingCity)) #fronteir

    while not pq.empty():
        u = pq.get()[1]     
        ucs_traversal_output.append(u)
        if u== destinationNode:
            break
        visited[u] = True
        for v in romanian_map[u].keys():
            if not visited[v]:                
                parent[v] = u
                distance[v] = distance[u] + romanian_map[u][v]
                pq.put((distance[v],v))

    g = destinationNode
    path = []
    while g is not None:
        path.append(g)
        g = parent[g]

    path.reverse()
    print(path)
    print(distance[destinationNode])
    print(ucs_traversal_output)


ucs('Arad', 'Bucharest')
```
