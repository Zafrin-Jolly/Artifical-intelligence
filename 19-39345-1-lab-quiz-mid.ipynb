```python
table={('B','Clean'):'Going to A',('B','Dirty'):'Suck',('C','Clean'):'Going to A',('C','Dirty'):'Suck',('A','Clean'):'Going to C',('A','Dirty'):'Suck'}
percepts=[]    
def table_driven_agent(percept):
    print(percept)
    percepts.append(percept)
    action = lookup(percept,table)
    return action

def lookup(percept,table):
    for key in table:
        if key == percept:
            return table[key]

import random
Location = random.choice(['A','B','C']) #randomly choosing the location and condition 
Condition = random.choice(['Clean','Dirty'])

while True:
    action= table_driven_agent((Location,Condition))#choosing action based of location and condition
    print(action)
    cmd = input()
    if(cmd == 'stop'): break
    if action == 'Goint to A': 
        Location = 'A' #got the location for the particular action
        Condition = random.choice(['Clean','Dirty'])
    elif action== 'Going to B':
        Location = 'B'#got the location for the particular action
        Condition = random.choice(['Clean','Dirty'])
    elif action== 'Going to C':
        Location = 'C'#got the location for the particular action
        Condition = random.choice(['Clean','Dirty'])
    else: 
        Condition = 'Clean'
```
