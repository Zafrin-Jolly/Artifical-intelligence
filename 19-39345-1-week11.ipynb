```python
cgraph = {'WA': ['SA', 'NT'], 'Q': ['SA', 'NT', 'NSW'], 'T': [],
'V': ['SA', 'NSW'], 'SA': ['WA', 'NT', 'Q', 'NSW','V'], 
          'NT': ['Q', 'WA', 'SA'], 'NSW': ['Q', 'V', 'SA']}
okColor={}
```

```python
# write a method to get the highest degree node

def get_node_hd():
    mnode = ''
    length = 0
    for k in cgraph.keys():
        if(len(cgraph[k])>length and k not in okColor.keys()):
            length = len(cgraph[k])
            mnode = k
    return mnode
```

```python
colorstochooselist={'WA': ['red','green', 'blue'], 
                    'Q': ['red','green', 'blue'],
                    'T': ['red', 'green', 'blue'],
                    'V':  ['red','green', 'blue'], 
                    'NT': ['red','green', 'blue'],
                    'SA': ['red','green', 'blue'],                    
                    'NSW': ['red','green', 'blue']}

import random
def get_color(node):
    # check colorstochooselist[node] is empty or not
    # return 'no color'
    if len(colorstochooselist[node]) == 0:
        return 'no color'
    color = random.choice(colorstochooselist[node])
    colorstochooselist.pop(node)
    # update color options
    adjList = cgraph[node]
    for adj in adjList:
        if(adj in colorstochooselist.keys()):
            colorstochooselist[adj].remove(color)
    #print(adjList)
    return color
    
```

```python
# assign color to the highest degree node
hdNode = get_node_hd()
okColor[hdNode]= get_color(hdNode)
print(colorstochooselist)
print(okColor)
```

```python
# define a method to select a node by MRV
def get_node_MRV():
    min_value= min([len(colorstochooselist[ele]) for ele in colorstochooselist])
    for ele in colorstochooselist:
        if len(colorstochooselist[ele]) == min_value:
            node=ele
       # node= colorstochooselist.index(min_value)
    #this will give the minimum length of the list in the dictionary
    return node
```

```python
L = len(colorstochooselist)
print("the First colored state is:",okColor,"\n")
for i in range(0,L):
    mrvNode = get_node_MRV()
    # assign color to the Minimum Remaining Value
    selected_color= get_color(mrvNode)
    # if no color returned then print failure
    if (selected_color=='no color'):
        print("Failure")
        break
    okColor[mrvNode]= selected_color
    print("Remaining :",colorstochooselist)
    print("the colored states :",okColor,"\n")
print("\nAll state colored!")
```
