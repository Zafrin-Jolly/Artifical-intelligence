# N Queen Attack Detection

```python
import random
N = 4
r = []
c = []
for i in range(1, N+1):
    r.append(random.choice(range(1, N+1)))
    c.append(random.choice(range(1, N+1)))
    print('Q{}: {},{}'.format(i,r[-1],c[-1]))
print(r)
print(c)
```

```python
# row attack
for i in range(0,N):
    for j in range(i+1,N):
        if(r[i] == r[j]):
            print('row attack: row:{} column:{} and {}'.format(r[i],c[i],c[j]))      
       
```

```python
#column attack
for k in range(0,N):
    for l in range(k+1,N):
        if(c[k] == c[l]):
            print('column attack: column:{} row:{} and {}'.format(c[k],r[k],r[l]))  
```

```python
# diagonal atack
for i in r:
    for k in r:
        deltaRow = abs(r[i] -r[k])
for j in c:
    for l in c:
        deltaCol = abs(c[j] - c[l])
if(deltaRow==deltaCol):
    print('Diagonal attack between row:{} column:{} and row:{} column:{}'.format(r[i],c[j],r[k],c[l]))
       
```
